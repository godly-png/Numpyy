```python
df=pd.read_csv("C:\\Users\\godly\\Downloads\\StudentsPerformance.csv")
```

```python
df['gender'].replace({'male': 0, 'female': 1}, inplace=True)

df['test preparation course'].replace({'none': 0, 'completed': 1},inplace=True)
df
```

```python
# • Compute average and standard deviation of scores by groups—e.g., by gender, parental
# education, preparation status (use np.std)

import pandas as pd
import numpy as np
mean_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
std_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].std()
print("Mean scores by gender:\n", mean_scores)
print("\nStandard deviation by gender:\n", std_scores)


```

```python
# • Explore score distributions and inter-subject correlations using NumPy
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\godly\\Downloads\\StudentsPerformance.csv")
corr_df = df[['math score', 'reading score', 'writing score']].corr()
print(corr_df)




```

```python
# Bar charts: Average scores by parental education or test-prep status




import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\godly\\Downloads\\StudentsPerformance.csv")
sns.barplot(data=df, x='parental level of education', y='math score')
plt.title('Average Math Score by Parental Education')
plt.show()
sns.barplot(data=df, x='test preparation course', y='math score')
plt.title("Average Scores by Test Preparation Course")
plt.show()
```

```python
# • Box plots: Score distributions grouped by gender or lunch type


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\godly\\Downloads\\StudentsPerformance.csv")
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Score Distribution by Gender')
plt.show()
sns.boxplot(x='lunch', y='math score', data=df)
plt.title('Math Score Distribution by Lunch Type')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\godly\\Downloads\\StudentsPerformance.csv")
sns.scatterplot(x='reading score', y='writing score', data=df)
plt.title('Reading vs Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```
